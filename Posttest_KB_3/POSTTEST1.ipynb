Informasi yang didapatkan adalah jumlah record, nama attribute, jumlah attribute dengan tipe objek, jumlah attribute dengan tipe angka, jumlah attribute yang memiliki nilai null.

```python
dataframe = pd.read_csv("hotstar.csv")
dataframe
```

Masukkan informasi di atas dalam dictionary dengan key sesuai nama informasinya. Tampilkan isi dari dictionary tersebut sebagus mungkin.

```python
types = dataframe.dtypes
columns_df = dataframe.columns
```

```python
columns_object = []
for index, item in enumerate(types):
    if str(item) == 'object':
        columns_object.append(columns_df[index])

df_object = dataframe[columns_object].dtypes
```

```python
columns_float = []
for index, item in enumerate(types):
    if str(item) == 'float64':
        columns_float.append(columns_df[index])

df_float = dataframe[columns_float].dtypes
```

```python
float64 = len(df_float)
```

```python
columns_int = []
for index, item in enumerate(types):
    if str(item) == 'int64':
        columns_int.append(columns_df[index])
        
df_int = dataframe[columns_int].dtypes
```

```python
int64 = len(df_int)
```

```python
jumlah_angka = float64 + int64
jumlah_angka
```

```python
record = len(dataframe)
atribut = dataframe.columns
objek = len(df_object)
angka = jumlah_angka
null = len(dataframe.columns[dataframe.isna().sum() > 0])
```

```python
dictionary = {
    "Jumlah record" : record,
    "Nama atribut"  : atribut,
    "Jumlah atribut tipe objek" : objek,
    "Jumlah atribut tipe angka" : angka,
    "Jumlah atribut nilai null" : null,
}
```
