```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('hotstar.csv')
dataset
```

## Grafik Batang

```python
plt.figure(figsize=(10,10), dpi=300)
plt.bar(dataset['year'], dataset['seasons'], 
 color = 'brown', align ='center')
plt.title("Perbandingan year dengan seasons", loc='left')
plt.xlabel("Year")
plt.ylabel("Seasons")
plt.figure(facecolor='black')
plt.show()
```

grafik batang ini membandingkan dua atribut yaitu atribut year dan seasons, dimana tiap tahun nya seasons hotstar akan naik turun dan jumlah season tiap tahun nya tidak menentu

## Grafik Plot dan Baris

```python
tipe = [item for item in range (0, len(dataset['type']))]
plt.figure(figsize=(20,10), dpi=300)
plt.plot(tipe, dataset['age_rating'], 'orange',
 'b', linewidth=3)
plt.plot(tipe, dataset['year'], 'red',
 'r', linestyle = 'dashed')
plt.plot(tipe, dataset['episodes'], 'brown',
 'g', marker ='o', markersize = 5)
plt.legend(['Age Rating', 'Year', 'Episodes'], loc = 'upper right')
plt.title('Data Hotstar')
plt.xlabel('Type')
plt.ylabel('Atribut')
plt.show()
```

grafik ini membandingkan atribut type dengan 3 atribut lainnya yaitu atribut year, age rating, dan seasons
dimana atribut age rating ini tidak memiliki perbedaan tiap type nya, atribut year hanya mengalami sedikit perbedaan dengan atribut type, dan atribut episodes memiliki perubahan yang signifikan atau mengalamai perubahan yang naik turun dengan atribut type

## Grafik Pie

```python
dataset['type'].value_counts()
```

```python
labels = dataset['type'].value_counts().rename_axis('Tipe').reset_index(name = 'Jumlah')
```

```python
plt.figure(figsize=(10,10), dpi=300)
plt.pie(dataset['type'].value_counts(),
 labels = labels['Tipe'], autopct = '%.1f%%', startangle = 100)
plt.title('Data Hotstar Atribut Type')
plt.show()
```

Grafik ini membandingkan record yang ada di dalam atribut, di atribut type membandingkan dua record dimana record movie lebih banyak dibandingkan record tv
